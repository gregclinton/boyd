```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

```python
X = []

A = np.array([[4, 3], [1, 2]])
b = np.array([1, 2])

def near(a, b):
    return np.linalg.norm(a - b) < 500

for x1 in np.linspace(0, 100, 101): 
    for x2 in np.linspace(0, 100, 101): 
        x = [x1, x2]
        if near(A.dot(x), b):
            X.append(x)

X = np.array(X)
plt.scatter(X[:, 0], X[:, 0], s = 0.8, edgecolors = 'none')
plt.show()
```
