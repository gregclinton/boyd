```python
X = []

a = np.array([14, 14])
b = 2000

gridx = gridy = np.linspace(0, 100, 101)

for x1 in gridx: 
    for x2 in gridy: 
        x = [x1, x2]
        if a.dot(x) < b:
            X.append(x)

X = np.array(X)
plt.figure(figsize = (4, 4))
plt.xlim(gridx[0], gridx[-1])
plt.ylim(gridy[0], gridy[-1])
plt.scatter(X[:, 0], X[:, 1], s = 0.8, edgecolors = 'none')
x0 = b * a / a.dot(a)
plt.plot(x0[0], x0[1], 'o')
plt.show()
```
